## Insertion Sort Algorithm
Python code to sort an array using InsertionSort

```python
array = [40,30,20,10]


def sorting(arr):
    length = len(arr)
    for i in range(1,length):
        key = arr[i]
        j = i-1 

        while j>=0 and key < arr[j]:
            arr[j+1] = arr[j]
            j-=1

        arr[j+1] = key
    
    return arr

sorted_array = sorting(array)
print(sorted_array)
```
